```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib widget
```

## Single field

```python
print("Parameters:")
print("T = 20")
print("dx = 0.01\n\n")

times = np.array([222.4, 137.15, 118.92, 104.46, 114.21])
n_threads = [1, 2, 4, 8, 16]
speedUp = times[0] / times

N_iter = 200000

print("Mean time per iteration:")
print(f"Serial   :   {times[0]/N_iter} s/iter")
print(f"2 threads:   {times[1]/N_iter} s/iter")
print(f"4 threads:   {times[2]/N_iter} s/iter")
print(f"8 threads:   {times[3]/N_iter} s/iter")
print(f"16 threads:  {times[4]/N_iter} s/iter")

fig, axs = plt.subplots(1,2, figsize=(13, 5))

axs[0].plot(n_threads, times, 'o--', color = 'r')
axs[0].plot(4, 89, 'o', color = 'b', label = 'TDVRK - all steps') # Valore "medio"
axs[0].plot([4,8], [75.18,69.96], 'o', color = 'g', label = 'TDVRK - all steps + fprime')

axs[0].set_xlabel("N threads")
axs[0].set_ylabel("Time [s]")

axs[0].legend()

axs[1].plot(n_threads, speedUp, 'o--', color = 'r')
axs[1].plot(4, times[0]/89, 'o', color = 'b') # Valore "medio"
axs[1].plot([4,8], [times[0]/75.18,times[0]/69.96 ],'o', color = 'g')

axs[1].set_xlabel("N threads")
axs[1].set_ylabel("Speed Up")
axs[1].set_title("SpeedUp = SerialTime / ParallelTime")

print("\n\nCPU 8 cores - 16 threads")
```
