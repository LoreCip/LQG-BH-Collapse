```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib widget
def central(i, dx, arr):
    return (arr[i+1] - arr[i-1]) / (2*dx)

def backward(i, dx, arr):
    return (arr[i-2] - 4*arr[i-1] + 3*arr[i]) / (2*dx)

def forward(i, dx, arr):
    return (-arr[i+2] + 4*arr[i+1] - 3*arr[i]) / (2*dx)
```

```python
i = 3

x = np.array([0.5, 1, 1.5, 2, 2.5, 3, 3.5])
data = np.array([0.5**2, 1, 1.5**2, 4, 4, 4, 4])

dc = central (i, 0.5, data)
df = forward (i, 0.5, data)
db = backward(i, 0.5, data)

print("Backward: ", db)
print("Central : ", dc)
print("Forward : ", df)

fig = plt.figure()
plt.plot(x, data, 'o-')
plt.plot(x[i], data[i], 'o', color = 'r')

ddd = []
for i in [1, 2, 3, 4, 5]:
    ddd.append(central(i, 0.5, data))

plt.plot(x[1:-1], ddd, 'o-')
```

```python
x, h = np.linspace(0, 10, 10000, retstep=True)
data = x**2
data[x >= 2] = 5**2

fig = plt.figure()
plt.plot(x, data, 'o-')

ddd = []
for i in range(1, len(x)-1):
    ddd.append(central(i, h, data))
fig = plt.figure()
plt.plot(x[1:-1], ddd)
```
